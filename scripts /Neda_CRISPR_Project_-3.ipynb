```python
#this data shows the genotypes that are present as well as the comparison between the alternate mouse genome and the reference genome
import cyvcf2
import pandas as pd

vcf_path = 'filtered_variants_chr2.vcf'

# Converting VCF to TSV
with open('output.tsv', 'w') as file:
    for variant in cyvcf2.VCF(vcf_path):
        genotype = variant.gt_types[0]
        data_string = f"{variant.CHROM}\t{variant.POS}\t{variant.ID}\t{variant.REF}\t{','.join(variant.ALT)}\t{variant.QUAL}\t{variant.FILTER}\t{genotype}\n"
        file.write(data_string)

# Load TSV file into New DataFrame
columns = ['CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'GENOTYPE']
df = pd.read_csv("output.tsv", delimiter="\t", header=None, names=columns)
df
```

```python
#converts the 'QUAL' column in a pandas DataFrame df to numeric values, filters the DataFrame to include only rows where the 'QUAL' value is greater than 60, and displays the filtered DataFrame
#table shows the filtered dataframe from the previous table

df['QUAL'] = pd.to_numeric(df['QUAL'], errors='coerce')

filtered_df = df[df['QUAL'] > 60]
filtered_df
```

```python
#Identifying mutations caused by CRISPR:
##table shows the filtered dataframe with mutations and the quality of the reads as well as the position of mutations

intended_sites = [36937210, 36996899, 85400441, 85776687, 85918029, 86198668, 86236802, 86658391, 87049235]

intended_mutations_df = filtered_df[df['POS'].isin(intended_sites)]
intended_mutations_df
```

```python
#Detecting Off-Target Effects caused by the CRISPR:
#table shows all off target effects created and the position they are located
off_target_mutations_df = filtered_df[~df['POS'].isin(intended_sites)]
off_target_mutations_df
```

```python
#Summarizing Off-Target Mutations
summary_df = off_target_mutations_df.groupby('CHROM')['POS'].agg(['count', 'nunique'])
summary_df.columns = ['Mutation Count', 'Genotype Count']
summary_df
```

```python
#creates a graph with the following qualifications
import matplotlib.pyplot as plt

plt.hist(off_target_mutations_df['POS'], bins=50)
plt.xlabel('Position in Genome')
plt.ylabel('Number of Mutations')
plt.title('Off-Target Mutations on Chromosome 2')
plt.show()
```

```python
#creates a new DataFrame by filtering out rows from df where the 'POS' column values match any value in the intended sites list
off_target_mutations_df = filtered_df[~df['POS'].isin(intended_sites)]
off_target_mutations_df
```

```python
#groups the off target mutations DataFrame by the 'CHROM' column, calculates the total count of mutations and the counts of each genotype
off_target_summary_df = off_target_mutations_df.groupby('CHROM').agg(
    mutation_count=('POS', 'count'),
    genotype_counts=('GENOTYPE', 'value_counts'))
off_target_summary_df
```
